```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
#from adspy_shared_utilities import (
#plot_class_regions_for_classifier_subplot)
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

#Doing some data wrangling, erasing columns not relevant and categorizing string features into floats
df_train = df_train.drop(['Name','Ticket','Cabin'],axis=1).replace({'Sex': {'male': 1, 'female': 0}}).replace({'Embarked': {'S': 0, 'C': 1, 'Q': 2}})
df_train = df_train.dropna()
df_test = df_test.drop(['Name','Ticket','Cabin'],axis=1).replace({'Sex': {'male': 1, 'female': 0}}).replace({'Embarked': {'S': 0, 'C': 1, 'Q': 2}}).dropna()

#Defining X_train and y_train
X_train = df_train.drop('Survived',axis=1)
y_train = df_train['Survived']

clf = LogisticRegression(C=100).fit(X_train, y_train)
print('the score of this logistic regression model is: ', clf.score(X_train, y_train))
```

```python
prediction = df_test.copy()

prediction['Survived'] = clf.predict(df_test)

prediction1 = prediction[['PassengerId','Survived']]
prediction
```

```python
prediction1.to_csv('Submission.csv',index=False)
```
